```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
x = np.linspace(0, 20, 100)
```

```python
red_f = lambda x: 100 * np.exp(-0.5*x)
yellow_f = lambda x: 100 * np.exp(-0.12*x)
blue_f = lambda x: 100 * np.exp(-0.05*x)
green_f = lambda x: 100 * np.exp(-0.055*x)
```

```python
red_y = red_f(x)
yellow_y = yellow_f(x)
blue_y = blue_f(x)
green_y = green_f(x)
```

```python
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)
plt.figure(dpi=300)
ax = plt.subplot(111)
ax.plot(x, red_y, label='Red')
ax.plot(x, blue_y, linestyle=':', label='Blue')
ax.plot(x, green_y, linestyle='-.', label='Green')
# ax.plot(x, yellow_y, color='yellow', label='Yellow')
ax.legend(loc='upper right')
plt.xlim((0,20))
plt.ylim((0,100))
plt.ylabel('(%)', fontproperties=font)
plt.xlabel('(m)', fontproperties=font)
plt.show()
```
