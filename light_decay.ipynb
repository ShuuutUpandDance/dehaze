```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
```

```python
red_f = lambda x: 100 * np.exp(-0.5*x)
yellow_f = lambda x: 100 * np.exp(-0.12*x)
blue_f = lambda x: 100 * np.exp(-0.05*x)
green_f = lambda x: 100 * np.exp(-0.055*x)
```

```python
red_y = red_f(x)
yellow_y = yellow_f(x)
blue_y = blue_f(x)
green_y = green_f(x)
```

```python
plt.figure()
ax = plt.subplot(111)
ax.plot(x, red_y, color='red', label='Red')
ax.plot(x, blue_y, color='blue', label='Blue')
ax.plot(x, green_y, color='green', label='Green')
ax.plot(x, yellow_y, color='yellow', label='Yellow')
# ax.legend(loc='center left', bbox_to_anchor=(0.5, -0.05),  shadow=True, ncol=2)
plt.xlim((0,20))
plt.ylim((0,100))
plt.ylabel('Light Intensity(%)')
plt.xlabel('Propagation Distance(m)')
plt.show()
```
